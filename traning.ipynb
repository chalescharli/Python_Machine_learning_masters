```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Create a DataFrame with the data
data = pd.DataFrame({'Study Time': [1, 2, 3, 4, 5], 'Test Scores': [60, 70, 75, 80, 85]})
```

```python
data
```

```python
# Extract features (study time) and target variable (test scores)
X = data[['Study Time']]
y = data['Test Scores']
```

```python
# Create a linear regression model
model = LinearRegression()
```

```python
# Fit the model to the data
model.fit(X, y)
```

```python
X
```

```python
# Predict test scores for new study time values
new_study_time = [[6],[7],[0]]  # Example new study time values
predicted_scores = model.predict(new_study_time)
print(predicted_scores)
```

```python
# Visualize the relationship between study time and test scores
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.scatter(new_study_time, predicted_scores, color='green', label='Predicted Scores')
plt.xlabel('Study Time')
plt.ylabel('Test Scores')
plt.title('Study Time vs Test Scores')
plt.legend()
plt.show()
```

```python
# Print the coefficients of the linear regression model
print('Intercept:', model.intercept_)
print('Coefficient:', model.coef_)
```
